```python
import os
import requests
import shutil
from config.paths import DJANGO_SQLITE_DB_PATH, CHROMA_SQLITE_PATH, TMP_DIR

def download_file_from_google_drive(file_id, destination):
    URL = f"https://drive.usercontent.google.com/download?id={file_id}&export=download&confirm=t&uuid=6315cf53-3f01-4acb-b286-094e0d0b48e2"
    session = requests.Session()

    def get_confirm_token(response):
        for key, value in response.cookies.items():
            if key.startswith('download_warning'):
                return value
        return None

    def save_response_content(response, destination):
        CHUNK_SIZE = 32768

        with open(destination, "wb") as f:
            for chunk in response.iter_content(CHUNK_SIZE):
                if chunk:
                    f.write(chunk)

    response = session.get(URL, params={'id': file_id}, stream=True)
    token = get_confirm_token(response)

    if token:
        params = {'id': file_id, 'confirm': token}
        response = session.get(URL, params=params, stream=True)

    save_response_content(response, destination)



chroma_db_id = '1mpOiaBOk4XqGfQ8DwyyKSZNrOsyMYH3J'
django_db_id = '1vGsrE0TRiv39z7jROAknXk4x1CS96Hdz'

tmp_chroma_db_path = os.path.join(TMP_DIR, "chroma.zip")
tmp_django_db_path = os.path.join(TMP_DIR, "django_db.zip")

if not os.path.exists(TMP_DIR):
    os.makedirs(TMP_DIR)

if not os.path.exists(CHROMA_SQLITE_PATH):
    download_file_from_google_drive(chroma_db_id, tmp_chroma_db_path)
    shutil.unpack_archive(tmp_chroma_db_path, os.path.dirname(CHROMA_SQLITE_PATH))


if not os.path.exists(DJANGO_SQLITE_DB_PATH):
    download_file_from_google_drive(django_db_id, tmp_django_db_path)
    shutil.unpack_archive(tmp_django_db_path, os.path.dirname(DJANGO_SQLITE_DB_PATH))

print("Databases downloaded and extracted successfully!")
```
